```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('median_ndvi_20per_cc_clean.csv')
```

```python
sns.catplot(data=df, x='month', y='NDVI', kind='bar', col="year", col_wrap=3, color='steelblue') 
plt.title('Median NDVI per Year') 
plt.xlabel('Month') 
plt.ylabel('Median NDVI') 
plt.show()
```
